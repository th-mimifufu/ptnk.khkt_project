## 1. Generate synthetic test data

```python
test_df = pd.read_excel("../data/new_l2.xlsx")
mapping = {
    8.0: 1,
    7.0: 2,
    6.5: 3,
    5.0: 4
}
test_df["hl12"] = test_df["hl12"].astype(float)
test_df['hl12'] = test_df['hl12'].map(mapping).fillna(test_df['hl12'])

test_df.loc[test_df['hl12'] == 8.0, 'hl12'] = 1
test_df.loc[test_df['hl12'] == 7.0, 'hl12'] = 2
test_df.loc[test_df['hl12'] == 6.5, 'hl12'] = 3
test_df.loc[test_df['hl12'] == 5.0, 'hl12'] = 4

test_df["hl12"] = test_df["hl12"].astype(int)

bins = [0, 5_000_000, 10_000_000, 20_000_000, 40_000_000, 60_000_000, 80_000_000, 100_000_000, float('inf')]
labels = ["0-5M", "5M-10M", "10M-20M", "20M-40M", "40M-60M", "60M-80M", "80M-100M", "Above 100M"]
test_df['tuition_range'] = pd.cut(test_df['Học phí'], bins=bins, labels=labels, right=False)
```

```python
test_df[test_df['Tổ hợp môn'] =='HocBa']
```

```python
# Groupby 'Tổ hợp môn' and 'ma_xet_tuyen' lấy the row that have the lowest 'Điểm chuẩn' of each group
grouped = test_df.groupby(['Tổ hợp môn', 'ma_xet_tuyen']).aggregate({
    'Điểm chuẩn': 'min',
    'Học phí': 'first',
    'tuition_range': 'first',
    'Công lập (1/0)': 'first',
    'Tỉnh/TP': 'first',
    'tên ccta': 'first',
    'điểm ccta': 'first',
    'điểm quy đổi': 'first',
    'hk10': 'first',
    'hk11': 'first',
    'hk12': 'first',
    'hl10': 'first',
    'hl11': 'first',
    'hl12': 'first',
    'nhom_nganh': 'first'
}).reset_index()
```

```python
grouped[grouped['Tổ hợp môn'] == 'HocBa']
```

```python
import random

thpt_incs = [random.uniform(0.1, 5) for _ in range(10)]
vsat_incs = [random.uniform(1, 100) for _ in range(10)]
dgnl_incs = [random.uniform(1, 300) for _ in range(10)]
hocba_incs = [random.uniform(0.1, 5) for _ in range(10)]

rows = []
for _, row in grouped.iterrows():
    base = row['Điểm chuẩn']
    code = str(row['ma_xet_tuyen'])
    
    # 1. Thêm row gốc với label "origin"
    origin = row.copy()
    origin['synthetic'] = 'origin'
    rows.append(origin)
    
    # 2. Tạo 5 synthetic rows
    for _ in range(5):
        if base == -1:
            new_score = -1
            label = 'origin'
        else:
            # Chọn danh sách và limit
            if "THPT" in code:
                inc_list, limit = thpt_incs, 30
            elif "VSAT" in code:
                inc_list, limit = vsat_incs, 450
            elif "ĐGNL" in code:
                inc_list, limit = dgnl_incs, 1200
            elif "HocBa" in code:
                inc_list, limit = hocba_incs, 30
            else:
                inc_list, limit = [], None
            
            if not inc_list:
                new_score, label = base, 'origin'
            else:
                inc = random.choice(inc_list)
                sign = random.choice([-1, 1])  # 50% cộng, 50% trừ
                candidate = base + sign * inc
                # Kiểm tra limit và không bé hơn 0
                if limit and (candidate > limit or candidate < 0):
                    # Nếu candidate vi phạm, thử chiều ngược lại
                    sign *= -1
                    candidate = base + sign * inc
                    if limit and (candidate > limit or candidate < 0):
                        candidate = base     # không thay đổi nếu vẫn vi phạm
                        sign = 0
                new_score = candidate
                label = 'add' if sign > 0 else ('minus' if sign < 0 else 'origin')
        
        new_row = row.copy()
        new_row['Điểm chuẩn'] = new_score
        new_row['synthetic'] = label
        rows.append(new_row)

# Chuyển thành DataFrame
synthetic_test_df = pd.DataFrame(rows)

# Kiểm tra kết quả
print(synthetic_test_df['synthetic'].value_counts())
synthetic_test_df.head(10)
```

### 2. Find list of ma_xet_tuyen_true_list (những trường-ngành-ptxt còn lại phù hợp với features)

```python
import polars as pl

# 0) Copy test DataFrame
test_pd = synthetic_test_pos_not_label.copy()

# 1) Hard-rename giống train
rename_map = {
    "Công lập (1/0)": "cong_lap",
    "Điểm chuẩn":        "diem_chuan",
    "điểm quy đổi":      "diem_quy_doi",
    "Tỉnh/TP":           "tinh_tp",
    "Tổ hợp môn":        "to_hop_mon",
    "tên ccta":          "ten_ccta",
    "điểm ccta":         "diem_ccta",
}
test_pd.rename(columns=rename_map, inplace=True)

# 2) Đảm bảo categorical là string
categorical_features = ["tuition_range","tinh_tp","to_hop_mon","ten_ccta","diem_ccta"]
for c in categorical_features:
    test_pd[c] = test_pd[c].astype(str)

# 3) Convert sang Polars
df = pl.from_pandas(test_pd)

# 4) Thêm idx giữ vị trí
df = df.with_row_count("idx")

# 5) Cross-join & filter
df_cross = df.join(df, how="cross", suffix="_r")
cond = (
    (pl.col("diem_chuan_r") >= pl.col("diem_chuan")) &
    (pl.col("to_hop_mon_r") == pl.col("to_hop_mon")) &
    (pl.col("Học phí_r") >= pl.col("Học phí")) &
    (pl.col("tinh_tp_r") == pl.col("tinh_tp")) &
    (pl.col("hk10_r") <= pl.col("hk10")) &
    (pl.col("hk11_r") <= pl.col("hk11")) &
    (pl.col("hk12_r") <= pl.col("hk12")) &
    (pl.col("hl10_r") <= pl.col("hl10")) &
    (pl.col("hl11_r") <= pl.col("hl11")) &
    (pl.col("hl12_r") <= pl.col("hl12"))
)
filtered = df_cross.filter(cond)
```

```python
filtered
```

```python
grouped = (
    filtered
      .group_by("idx")
      .agg(
        pl.col("ma_xet_tuyen_r")
          .implode().alias("ma_xet_tuyen_true_list"),
      )
)

result = (
    df
      .join(grouped, on="idx", how="left")
      .drop("idx")
)
```

### 3. Final Result

```python
result
```

```python
check = ["THPT", "VSAT"]

for c in check:
    
    exploded = result.explode("ma_xet_tuyen_true_list")

    # Lọc: Nếu ma_xet_tuyen chứa c, chỉ giữ dòng nào ma_xet_tuyen_true_list cũng chứa c,
    # còn lại giữ nguyên (hoặc None cũng được, sẽ bị bỏ khi group lại)
    filtered = exploded.filter(
        (
            pl.col("ma_xet_tuyen").str.contains(c)
            & pl.col("ma_xet_tuyen_true_list").str.contains(c)
        ) | (
            ~pl.col("ma_xet_tuyen").str.contains(c)
        )
    )

    
    all_cols = [col for col in result.columns if col != "ma_xet_tuyen_true_list"]
    result = (
        filtered.group_by(all_cols)
        .agg(pl.col("ma_xet_tuyen_true_list")
          .implode().alias("ma_xet_tuyen_true_list"))
    )
```
