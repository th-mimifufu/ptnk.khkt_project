```python
import re
TOKEN_MAP = {
    "NTS01": NTS01,
    "NTS02": NTS02
}

TOKEN_PATTERN = re.compile("|".join(map(re.escape, TOKEN_MAP.keys())))

def expand_ma(val: str):
    """Trả về list các giá trị ma_xet_tuyen sau khi thay token bằng từng phần tử trong list.
    Nếu không có token thì trả về [val] để giữ nguyên 1 dòng.
    """
    if pd.isna(val):
        return [val]
    val = str(val)
    m = TOKEN_PATTERN.search(val)
    if not m:
        return [val]  # không có token cần thay
    token = m.group(0)
    repl_list = TOKEN_MAP[token]
    return [val.replace(token, repl) for repl in repl_list]

df2 = df.copy()

df2["ma_xet_tuyen"] = df2["ma_xet_tuyen"].astype(str)

df2["ma_xet_tuyen"] = df2["ma_xet_tuyen"].apply(expand_ma)
df2 = df2.explode("ma_xet_tuyen", ignore_index=True)
```

```python
def add_diem_chuan_final(df: pd.DataFrame) -> pd.DataFrame:
    """
    Return a copy of df with:
      - y_base            = diem_chuan + diem_quy_doi   (điểm chuẩn gốc)
      - diem_chuan_final  = y_base - diem_quy_doi       (ngưỡng thật để so sánh)

    Assumptions:
      - For CCTA rows, recorded diem_chuan = base - diem_quy_doi.
      - For base rows, diem_quy_doi = 0.
    """
    
    out = df.copy()

    if 'diem_chuan' not in out.columns:
        raise ValueError("Missing 'diem_chuan' column in input DataFrame.")
    if 'diem_quy_doi' not in out.columns:
        out['diem_quy_doi'] = 0

    def _to_num(x):
        if pd.isna(x): return 0.0
        if isinstance(x, str):
            x = x.replace(',', '').replace('$', '').strip()
        try:
            return float(x)
        except Exception:
            return 0.0

    out['diem_chuan']   = out['diem_chuan'].map(_to_num)
    out['diem_quy_doi'] = out['diem_quy_doi'].map(_to_num)

    # Core logic
    out['y_base'] = out['diem_chuan']
    out['diem_chuan_final'] = out['diem_chuan'] - out['diem_quy_doi']

    if 'ten_ccta' in out.columns and 'diem_ccta' in out.columns:
        out['is_base_row'] = (
            (out['diem_quy_doi'] == 0) &
            (out['ten_ccta'].astype(str).isin(['0','', 'nan'])) &
            (out['diem_ccta'] == 0)
        )
    else:
        out['is_base_row'] = out['diem_quy_doi'] == 0

    base_miss = (out.loc[out['is_base_row'], 'diem_chuan_final'] - out.loc[out['is_base_row'], 'diem_chuan']).abs() > 1e-6
    if base_miss.any():
        print(f"[WARN] {int(base_miss.sum())} base rows have final != base (should be ==). Check input.")

    ccta_rows = out['diem_quy_doi'] > 0
    bad_ccta  = (out.loc[ccta_rows, 'diem_chuan_final'] >= out.loc[ccta_rows, 'diem_chuan'])
    if bad_ccta.any():
        print(f"[WARN] {int(bad_ccta.sum())} CCTA rows have final >= base. Verify 'diem_quy_doi' sign/values.")
            
    return out
```

```python
import numpy as np
import pandas as pd

def filter_candidates_per_student(
    student_df: pd.DataFrame,
    candidate_df: pd.DataFrame,
    neg_pos_ratio: int = 3,
    hard_filters=None,
    stu_budget_col: str = 'hoc_phi',
    cand_tuition_col: str = 'hoc_phi',
    default_budget: float = np.inf
) -> pd.DataFrame:
    """
    Build (user, candidate) pairs for LightGBM ranking.

    Assumes candidate_df already has 'diem_chuan_final' (true cutoff after CCTA fix).
    Uses student's budget = student_df[stu_budget_col] and candidate's tuition = candidate_df[cand_tuition_col].

    Adds features:
      - cand_margin      = student_diem_chuan - cand_diem_chuan_final
      - cand_budget_ok   = 1 if tuition <= budget
      - cand_budget_diff = budget - tuition
    """
    if hard_filters is None:
        # IMPORTANT: do NOT include any fee column here
        hard_filters = ['tinh_tp', 'to_hop_mon', 'cong_lap', 'nhom_nganh']

    if 'diem_chuan_final' not in candidate_df.columns:
        raise ValueError("candidate_df missing 'diem_chuan_final' (run CCTA normalization first)")

    # ----- copies & dtype normalization -----
    cand = candidate_df.copy()
    cand['hoc_phi'] = pd.to_numeric(cand[cand_tuition_col], errors='coerce') if cand_tuition_col in cand.columns else np.nan

    stu = student_df.copy()
    stu['budget_max'] = pd.to_numeric(stu[stu_budget_col], errors='coerce') if stu_budget_col in stu.columns else default_budget
    if 'diem_chuan' in stu.columns:
        stu['diem_chuan'] = pd.to_numeric(stu['diem_chuan'], errors='coerce').fillna(0)

    cand_records = cand.to_dict('records')  # each is a dict of scalars
    cand_cols = set(cand.columns)
    stu_cols  = set(stu.columns)

    rows = []
    for _, stu_row in stu.iterrows():
        budget = float(stu_row.get('budget_max', default_budget))
        if not np.isfinite(budget) or pd.isna(budget):
            budget = default_budget

        # ---- hard filters (no hoc_phi equality) ----
        filtered = cand_records
        for k in hard_filters:
            if (k in stu_cols) and (k in cand_cols):
                v = stu_row.get(k)
                filtered = [cr for cr in filtered if cr.get(k) == v]

        # ---- budget filter ----
        filtered = [cr for cr in filtered if pd.notna(cr.get('hoc_phi')) and float(cr['hoc_phi']) <= budget]
        np.random.shuffle(filtered)

        suitable, not_suitable = [], []
        stu_score = float(stu_row.get('diem_chuan', 0.0))

        for cand_row in filtered:
            cutoff_final = float(cand_row.get('diem_chuan_final', 0.0))
            margin = stu_score - cutoff_final

            tuition = float(cand_row.get('hoc_phi', np.nan))
            budget_ok  = int(pd.notna(tuition) and tuition <= budget)
            budget_diff = (budget - tuition) if (pd.notna(tuition) and np.isfinite(budget)) else np.nan

            # Optional: transcript constraints (compare scalar↔scalar only if both present)
            hb_cols = ['hk10','hk11','hk12','hl10','hl11','hl12']
            ok_hb = True
            for col in hb_cols:
                s_val = pd.to_numeric(stu_row.get(col, np.nan), errors='coerce')
                c_val = pd.to_numeric(cand_row.get(col, np.nan), errors='coerce')
                if pd.notna(s_val) and pd.notna(c_val) and s_val > c_val:
                    ok_hb = False
                    break

            # Final label
            is_match = (margin >= 0) and bool(budget_ok) and ok_hb

            row = {}
            row.update({f'student_{k}': v for k, v in stu_row.items()})
            row.update({f'cand_{k}':    v for k, v in cand_row.items()})
            row['cand_diem_chuan_final'] = cutoff_final
            row['cand_margin'] = margin
            row['cand_budget_ok'] = budget_ok
            row['cand_budget_diff'] = budget_diff
            row['label'] = int(is_match)

            (suitable if is_match else not_suitable).append(row)

        rows.extend(suitable)
        if not_suitable:
            k = min(len(suitable) * neg_pos_ratio, len(not_suitable))
            if k > 0:
                rows.extend(np.random.choice(not_suitable, size=k, replace=False).tolist())

    return pd.DataFrame(rows)
```

```python
import random
from typing import Optional

def generate_simple_synthetic(
    student_profile: pd.DataFrame,
    n_synthetic: int = 5,
    random_state: Optional[int] = 42,
    score_col: str = "diem_chuan",   # điểm của học sinh
    budget_col: str = "hoc_phi",     # ngân sách học sinh trong input ban đầu
) -> pd.DataFrame:
    """
    Tạo bản mở rộng student_profile gồm 1 dòng gốc + n_synthetic dòng nhiễu cho mỗi học sinh.
    - Giữ đúng thang điểm theo kỳ thi (THPT: 0–30, VSAT: 0–450, ĐGNL: 0–1200).
    - Đổi 'hoc_phi' của học sinh thành 'budget_max' (ngân sách), không nhầm với học phí của candidate.
    - Random chứng chỉ CEFR/JLPT như khả năng học sinh có.

    Trả về DataFrame có cột:
      - score_col (điểm HS sau nhiễu), budget_max (ngân sách), ten_ccta, diem_ccta, hl*/hk*, synthetic ∈ {origin, synthetic}
    """
    rng = random.Random(random_state)

    # Một ít bước nhiễu cho từng thang điểm (không cố định, chỉ là set tham khảo)
    thpt_incs = [rng.uniform(0.1, 5)  for _ in range(10)]
    vsat_incs = [rng.uniform(1, 100)  for _ in range(10)]
    dgnl_incs = [rng.uniform(1, 300)  for _ in range(10)]

    cefr_levels = ["A1", "A2", "B1", "B2", "C1", "C2"]
    jlpt_levels = ["N1", "N2", "N3", "N4", "N5"]

    def _num(x, default=0.0):
        try:
            return float(str(x).replace(",", "").replace("$", "").strip())
        except Exception:
            return default

    rows = []
    for _, row in student_profile.iterrows():
        code = str(row.get("ma_xet_tuyen", ""))

        # --- DÒNG GỐC ---
        origin = row.copy()
        # budget_max = ngân sách của học sinh (lấy từ budget_col)
        origin["budget_max"] = _num(origin.get(budget_col, 0))
        origin["synthetic"] = "origin"
        rows.append(origin)

        # --- DÒNG NHIỄU ---
        for _ in range(n_synthetic):
            new_row = row.copy()

            # 1) Điểm học sinh (score_col) – nhiễu theo kỳ thi
            val = _num(row.get(score_col, 0))
            if "THPT" in code:
                inc_list, lo, hi = thpt_incs, 0, 30
            elif "VSAT" in code:
                inc_list, lo, hi = vsat_incs, 0, 450
            elif "ĐGNL" in code or "DGNL" in code:
                inc_list, lo, hi = dgnl_incs, 0, 1200
            else:
                # Fallback: coi như THPT
                inc_list, lo, hi = thpt_incs, 0, 30

            if inc_list:
                inc = rng.choice(inc_list)
                sign = rng.choice([-1, 1])
                candidate = val + sign * inc
                # ép về miền hợp lệ
                if candidate < lo or candidate > hi:
                    candidate = val - sign * inc
                    if candidate < lo or candidate > hi:
                        candidate = max(lo, min(hi, val))
                new_val = candidate
            else:
                new_val = val
            new_row[score_col] = new_val

            # Ngân sách học sinh -> budget_max (KHÁC học phí trường)
            # Tạo nhiễu quanh ngân sách, giữ >= 0
            fee_val = _num(row.get(budget_col, 0))
            delta = rng.randint(10_000_000, 30_000_000)
            sign = rng.choice([-1, 1])
            fee_candidate = fee_val + sign * delta
            new_row["budget_max"] = max(0, fee_candidate)

            # Chứng chỉ ngoại ngữ mà học sinh có (tuỳ xác suất)
            ten_ccta = rng.choices(["CEFR", "JLPT", "0"], weights=[0.35, 0.25, 0.40])[0]
            if ten_ccta == "CEFR":
                diem_ccta = rng.choice(cefr_levels)
            elif ten_ccta == "JLPT":
                diem_ccta = rng.choice(jlpt_levels)
            else:
                ten_ccta = "0"
                diem_ccta = "0"
            new_row["ten_ccta"] = ten_ccta
            new_row["diem_ccta"] = diem_ccta

            # Điểm thành phần học bạ
            for col in ["hl10", "hl11", "hk10", "hk11", "hk12"]:
                new_row[col] = rng.choices([1, 2, 3, 4], weights=[0.4, 0.4, 0.1, 0.1])[0]
            new_row["hl12"] = rng.choices([1, 2, 3, 4, 5], weights=[0.3, 0.3, 0.3, 0.05, 0.05])[0]

            new_row["synthetic"] = "synthetic"
            rows.append(new_row)

    out = pd.DataFrame(rows)

    return out
```

```python
# student_budget_col làm tròn lên theo list [0, 10000000, 20000000, 40000000, 60000000, 80000000, 100000000, 200000000, 500000000]

def round_budget(x):
    if x <= 0:
        return 0
    elif x <= 10000000:
        return 10000000
    elif x <= 20000000:
        return 20000000
    elif x <= 40000000:
        return 40000000
    elif x <= 60000000:
        return 60000000
    elif x <= 80000000:
        return 80000000
    elif x <= 100000000:
        return 100000000
    elif x <= 200000000:
        return 200000000
    else:
        return 500000000

pairs['student_budget_max'] = pairs['student_budget_max'].apply(round_budget)
```

```python
import re, numpy as np, pandas as pd

# chọn cột ngân sách của student
student_budget_col = 'student_budget_max' if 'student_budget_max' in pairs.columns else 'student_hoc_phi'

# ép kiểu số cho các cột chính
for c in ['student_diem_chuan', 'cand_diem_chuan_final', student_budget_col, 'cand_hoc_phi']:
    if c in pairs.columns:
        pairs[c] = pd.to_numeric(pairs[c], errors='coerce')

# tính margin & budget
pairs['cand_margin'] = pairs['student_diem_chuan'] - pairs['cand_diem_chuan_final']
pairs['cand_budget_ok']   = (pairs['cand_hoc_phi'] <= pairs[student_budget_col]).astype('int8')
pairs['cand_budget_diff'] = (pairs[student_budget_col] - pairs['cand_hoc_phi'])

# chuẩn hóa học bạ: số nhỏ = tốt; 0/NaN => 10; candidate 10 = "không yêu cầu"
HB = ['hk10','hk11','hk12','hl10','hl11','hl12']
for col in HB:
    s_col = f'student_{col}'
    c_col = f'cand_{col}'
    if s_col in pairs.columns:
        pairs[s_col] = pairs[s_col].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
        pairs[s_col] = pairs[s_col].fillna(10).replace(0, 10)
    if c_col in pairs.columns:
        pairs[c_col] = pairs[c_col].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
        pairs[c_col] = pairs[c_col].fillna(10).replace(0, 10)

# hb_ok: tất cả yêu cầu đều thỏa: (cand==10) hoặc (student <= cand)
hb_ok = pd.Series(True, index=pairs.index)
for col in HB:
    s_col = f'student_{col}'
    c_col = f'cand_{col}'
    if s_col in pairs.columns and c_col in pairs.columns:
        hb_ok &= ((pairs[c_col] == 10) | (pairs[s_col] <= pairs[c_col]))
pairs['hb_ok'] = hb_ok.astype('int8')

# label cuối
pairs['label'] = ((pairs['cand_margin'] >= 0) & (pairs['cand_budget_ok'] == 1) & (pairs['hb_ok'] == 1)).astype('int8')
```
