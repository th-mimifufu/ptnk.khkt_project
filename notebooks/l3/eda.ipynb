```python
import pandas as pd
```

```python
df = pd.read_excel('../../data/L3.xlsx')
```

```python
df.info()
```

```python
df["truongnganhpt"].value_counts()
```

```python
# lây 3 kí tự đầu tiên của cột truongnganhpt làm mã trường
df['ma_truong'] = df['truongnganhpt'].str[:3]
```

```python
len(df['ma_truong'].unique())
```

```python
df['ma_truong'].value_counts()
```

```python
# Set lại học phí của trường QSA là 30 triệu
df.loc[df['ma_truong'] == 'QSA', 'Học phí'] = 30000000
```

```python
df[df["ma_truong"] == "QSA"]
```

```python
# drop row truongnganhpt = PVU7480101hb
df = df[df['truongnganhpt'] != 'PVU7480101hb']
```

```python
# fill NaN cho cột Học phí của trường DMD là 25 triệu
df.loc[(df['ma_truong'] == 'DMD') & (df['Học phí'].isna()), 'Học phí'] = 25000000
df.loc[(df['ma_truong'] == 'DVH') & (df['Học phí'].isna()), 'Học phí'] = 50000000
```

```python
df[df['ma_truong'] == 'DVH']
```

```python
df.to_excel('../../data/L3.xlsx', index=False)
```
